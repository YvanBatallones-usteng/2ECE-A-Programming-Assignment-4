```python
import pandas as pd 
df = pd.read_excel('board2.xlsx') # Load data from board2.xlsx into df
df
```

### Problem 1: Create the following data frames based on the format provided:

```python
df.loc[(df['Electronics'] >70)   #Filter rows where electronics score is greater than 70
        &(df['Track'] == 'Instrumentation') # and Track is Instrumentation 
        &(df['Hometown']=='Luzon'),  # and Hometown is Luzon
        ['Name', 'GEAS', 'Electronics'] #Select to display only Name, GEAS, and Electronics Columns
        ]
```

```python
df['Average'] = df[['GEAS', 'Electronics', 'Math']].mean(axis=1) # Compute row-wise the average scores of each sudents and stores in new columnn 
df
```

```python
df.loc[(df['Hometown'] == 'Mindanao')  # Filter rows where Hometown is Mindanao
        & (df['Gender'] == 'Female')  # and Gender is Female
        & (df['Average'] > 50),  # and Average score is greater than 50
        ['Name', 'Track', 'Electronics', 'Average']  # Select only Name, Track, Electronics, and Average columns
        ]
```

### Problem 2: Create a visualization that shows how the different features contributes to average grade

```python
import matplotlib.pyplot as plt #import matplotlib used for plotting

avg_by_track = df.groupby('Track')['Average'].mean() # Group data by Track and compute mean Average score

plt.bar(avg_by_track.index, avg_by_track.values) # Create a bar chart with Track on x-axis and Average scores on y-axis
plt.title("Averages by Track")
plt.ylabel("Average Score")
plt.xlabel("Track")
```

```python
avg_by_gender = df.groupby('Gender')['Average'].mean() # Group data by Gender and compute mean Average score

plt.bar(avg_by_gender.index, avg_by_gender.values) # Create a bar chart with Gender on x-axis and Average scores on y-axis
plt.title("Averages by Gender") 
plt.ylabel("Average Score")
plt.xlabel("Gender")
```

```python
avg_by_home = df.groupby('Hometown')['Average'].mean() # Group data by Hometown and compute mean Average score

plt.bar(avg_by_home.index, avg_by_home.values) # Create a bar chart with Hometown on x-axis and Average scores on y-axis
plt.title("Averages by Hometown")
plt.ylabel("Average Score")
plt.xlabel("Hometown")
```
